```python
# Переименовываем колонки
df2017.rename(columns={'Happiness.Score':'Happiness Score','Happiness.Rank':'Happiness Rank','Dystopia.Residual':'Dystopia Residual'},inplace=True)
df2018.rename(columns={'Country or region':'Country','Score':'Happiness Score','Overall rank':'Happiness Rank'},inplace=True)
df2019.rename(columns={'Country or region':'Country','Score':'Happiness Score','Overall rank':'Happiness Rank'},inplace=True)

# Добавляем год чтобы смочь различить 
dfs = [df2015,df2016,df2017,df2018,df2019]
year = 2015
for df in dfs:
    df['Year'] = year
    year+=1
    df["Median Score"] = df['Happiness Score'].median()    # Добавляем медиану
    df['Max score in year'] = df['Happiness Score'].max()    # Добавляем максимальное значение за год
    df['Min score in year'] = df['Happiness Score'].min()    #добавляем минимальное значение за год




# Добавление недостающей колонки
df2018['Dystopia Residual'] = df2018['Happiness Score'] -\
df2018['GDP per capita'] -\
df2018['Social support'] -\
df2018['Healthy life expectancy'] -\
df2018['Freedom to make life choices'] -\
df2018['Generosity'] -\
df2018['Perceptions of corruption']

df2019['Dystopia Residual'] = df2019['Happiness Score'] -\
df2019['GDP per capita'] -\
df2019['Social support'] -\
df2019['Healthy life expectancy'] -\
df2019['Freedom to make life choices'] -\
df2019['Generosity'] -\
df2019['Perceptions of corruption'] 

# Соединяем
df_score_of_year = pd.concat([df[['Year','Country',\
                                  'Happiness Score',\
                                  'Happiness Rank',\
                                  'Dystopia Residual',\
                                  "Median Score",\
                                  'Max score in year',\
                                  'Min score in year'\
                                 ]] for df in dfs],ignore_index=True)

# 
df_score_of_year
```

```python
# ДФ Постсоветского пространства по годам
df_ussr = df_score_of_year.loc[(df_score_of_year.Country =='Russia')|\
                               (df_score_of_year.Country =='Ukraine')|\
                               (df_score_of_year.Country =='Belarus')|\
                               (df_score_of_year.Country =='Georgia')|\
                               (df_score_of_year.Country =='Kazakhstan')|\
                               (df_score_of_year.Country =='Uzbekistan')|\
                               (df_score_of_year.Country =='Azerbaijan')|\
                               (df_score_of_year.Country =='Latvia')|\
                               (df_score_of_year.Country =='Estonia')|\
                               (df_score_of_year.Country =='Lithuania')|\
                               (df_score_of_year.Country =='Moldavia')|\
                               (df_score_of_year.Country =='Lithuania')|\
                               (df_score_of_year.Country =='Lithuania')|\
                               (df_score_of_year.Country =='Lithuania')|\
                               (df_score_of_year.Country =='Lithuania')|\
                               (df_score_of_year.Country =='Lithuania')|\
                               (df_score_of_year.Country =='Lithuania')]
df_ussr.head()
```
